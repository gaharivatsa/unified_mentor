```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\data analytics\greendestination.csv")
df
```

```python
df['Attrition'].value_counts('Yes')['Yes'] * 100
```

```python
df['Attrition_binary'] = df['Attrition'].apply(lambda x : 0 if x == 'No' else 1)
```

```python
plt.figure(figsize = (13,3))
```

```python
sns.countplot(data = df ,y = "Attrition" )
```

```python
plt.subplot(3,2,1)
sns.histplot(x = "Age" ,hue = "Attrition" , multiple='dodge',data = df,kde="True")

plt.subplot(3,2,2)
sns.histplot(x = "Age" ,hue = "Attrition" , multiple='stack',data = df,kde="True")



plt.tight_layout()
```

```python
data = df
```

```python
plt.subplot(3, 1, 1)
sns.histplot(data, x='Age', hue='Attrition', multiple='stack', palette='muted')
plt.title('Age vs. Attrition')

plt.subplot(3, 1, 2)
sns.histplot(data, x='YearsAtCompany', hue='Attrition', multiple='stack', palette='muted')
plt.title('Years at Company vs. Attrition')

plt.subplot(3, 1, 3)
sns.histplot(data, x='MonthlyIncome', hue='Attrition', multiple='stack', palette='muted')
plt.title('Monthly Income vs. Attrition')

plt.tight_layout()
```
